# Schleifen

### while Schleife

**Syntax**  
```java
while (ausdruck) {
    anweisung;
}
```
  
**Semantik**
```html
Ausdruck muss vom Typ boolean sein. Ist er true, wird die Anweisung ausgeführt, andernfalls mit der ersten Anweisung hinter der Schleife fortgefahren. 
```

**Beispiel**

````java
int zähler = 0;
while (zähler < 10) {
    System.out.println(zähler);
    zähler++;
}
````

### do Schleife

**Syntax**  
```java
do {
    anweisung;
} while (ausdruck);
```
  
**Semantik**
```html
Wird mindestens einmal ausgeführt: Die Anweisung wird 
ausgeführt und erst dann der Ausdruck ausgeführt. Falls der 
Ausdruck false ist, wird Ausführung abgebrochen.
```

**Beispiel**

````java
int zähler = 0;
do {
    System.out.println(zähler);
    zähler++;
} while (zähler < 10);
````

### for Schleife

**Syntax**  
```java
for (init; test; update) {
    anweisung;
} 
```
  
**Semantik**  
**init**-Ausdruck:  
* wird einmal vor Start durchlaufen
* kann Deklarationen und Initialisierungen von Zählervariablen enthalten
* kann auch entfallen
  
**test**-Ausdruck:
* Anweisung wird nur ausgeführt, wenn die Auswertung true ergibt

**update**-Ausdruck:
* Veränderung der Zählervariablen
* wird nach jedem Durchlauf der Schleife ausgewertet

**Beispiel**

````java
for (int i = 0;i < 10 ; i++) {
    System.out.println(i);
}
````
