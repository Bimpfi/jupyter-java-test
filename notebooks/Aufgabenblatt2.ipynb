### 2.2 Zahlensyteme

Schreiben Sie ein Konsolenprogramm Zahlensysteme, das jeweils einen Zahlenwert in andere Zahlensysteme konvertiert.  
Das folgende Beispiel zeigt die Benutzung anhand eines 
Beispiels:

```
Bitte geben Sie eine Zahl dezimal ein
11
Hex: b
Binär:1011
Bitte geben Sie eine Zahl hexadezimal ein
FF
Dezimal: 255
Binär: 11111111
Bitte geben Sie eine Zahl binär ein
101
Dezimal: 5
Hex: 5
Bitte geben Sie eine Zahl octal ein
77
Dezimal: 63
Hex: 3f

````java
import java.util.Scanner;
Scanner sc = new Scanner(System.in);

String input;

System.out.println("Bitte geben Sie eine Zahl dezimal ein.");
input = sc.nextLine();
System.out.println("Hex: " + Integer.toHexString(Integer.parseInt(input)));
System.out.println("Binär: " + Integer.toBinaryString(Integer.parseInt(input)));

System.out.println("Bitte geben Sie eine Zahl hexadezimal ein.");
input = sc.nextLine();
System.out.println("Dezimal: " + Integer.parseInt(input, 16));
System.out.println("Binär: " + Integer.toBinaryString(Integer.parseInt(input, 16)));

System.out.println("Bitte geben Sie eine Zahl binär ein.");
input = sc.nextLine();
System.out.println("Dezimal: " + Integer.parseInt(input, 2));
System.out.println("Hex: " + Integer.toHexString(Integer.parseInt(input, 2)));

System.out.println("Bitte geben Sie eine Zahl octal ein.");
input = sc.nextLine();
System.out.println("Dezimal: " + Integer.parseInt(input, 8));
System.out.println("Hex: " + Integer.toHexString(Integer.parseInt(input, 8)));
````
